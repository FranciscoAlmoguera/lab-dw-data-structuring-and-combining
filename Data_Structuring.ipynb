```python
import pandas as pd


url = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv"
data = pd.read_csv(url)


data.columns = data.columns.str.lower()
data.columns = data.columns.str.replace(' ', '_')
data.rename(columns={'st': 'state'}, inplace=True)

print("Cleaned column names:\n", data.columns)


data['gender'] = data['gender'].str.lower().replace({'f': 'F', 'm': 'M', 'femal': 'F', 'male': 'M'})


state_mapping = {
    'AZ': 'Arizona', 'Cali': 'California', 'WA': 'Washington'
}
import pandas as pd

# Load the original dataset
url1 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv"
url2 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv"
url3 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv"

# Load all datasets
data1 = pd.read_csv(url1)
data2 = pd.read_csv(url2)
data3 = pd.read_csv(url3)

# Function to clean and format a given dataframe
def clean_data(df):
    # Cleaning column names
    df.columns = df.columns.str.lower().str.replace(' ', '_').str.replace('st', 'state')

    # Cleaning invalid values
    df['gender'] = df['gender'].str.lower().replace({'f': 'F', 'm': 'M', 'femal': 'F', 'male': 'M'})
    df['state'] = df['state'].replace({'AZ': 'Arizona', 'Cali': 'California', 'WA': 'Washington'})
    df['education'] = df['education'].replace({'Bachelors': 'Bachelor'})
    df['customer_lifetime_value'] = df['customer_lifetime_value'].str.replace('%', '').astype(float)
    df['vehicle_class'] = df['vehicle_class'].replace({
        'Sports Car': 'Luxury', 'Luxury SUV': 'Luxury', 'Luxury Car': 'Luxury'
    })

    # Fixing data types
    df['customer_lifetime_value'] = pd.to_numeric(df['customer_lifetime_value'])
    df['number_of_open_complaints'] = df['number_of_open_complaints'].str.split('/').str[1].astype(int)

    return df

# Clean each dataset
clean_data1 = clean_data(data1)
clean_data2 = clean_data(data2)
clean_data3 = clean_data(data3)

# Combine the cleaned dataframes
combined_data = pd.concat([clean_data1, clean_data2, clean_data3], ignore_index=True)

# Save the combined cleaned dataset to a CSV file
combined_data.to_csv('combined_cleaned_data.csv', index=False)
print("Combined and cleaned dataset saved as 'combined_cleaned_data.csv'.")

# Load the marketing customer analysis dataset
marketing_url = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv"
marketing_data = pd.read_csv(marketing_url)

# Sales channel revenue analysis
sales_channel_pivot = marketing_data.pivot_table(
    values='total_claim_amount',
    index='sales_channel',
    aggfunc='sum'
).round(2)

print("Total revenue by sales channel:\n", sales_channel_pivot)

# Average customer lifetime value analysis per gender and education level
avg_clv_pivot = marketing_data.pivot_table(
    values='customer_lifetime_value',
    index=['gender', 'education'],
    aggfunc='mean'
).round(2)

print("Average customer lifetime value per gender and education level:\n", avg_clv_pivot)
```
